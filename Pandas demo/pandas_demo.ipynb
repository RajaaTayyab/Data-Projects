## Making my own Dataframe


```python
people = {
    "first" : ["Tayyab", "Nabeel" , "Armaghan"],
    "last" : ["Janjua" , "Babar" , "Awan"],
    "email" : ["tayyab@gmail.com", "nabeel@gmail.com", "armaghan@gmail.com"]
}
```

```python
df = pd.DataFrame(people) #makes dataframe
df #key becomes columns
```

```python
df["email"] #it is a series
```

```python
df.email  #gives same output either with dot or key 
#better practice is the key value one to avoid errors 
```

```python
type(df["email"]) #signular data #single output is caled series
```

```python
df[["last" , "email"]] 
```

```python
#we passed a list inside the brackets you need 2 brackets
#Series is single column of rows so it is not series
a=df[["last" , "email"]] 
type(a)
```

```python
df.columns #gives columns
```

```python
df.iloc[0] #gives first row
#rturns as sries of first row of data
```

```python
df.iloc[[0,1]] #first two rows now it returns datafrane
```

```python
df.iloc[[0,1], 2] #if we want email address of the first 2 rows
```

```python
df.loc[[0,1],]
```

```python
df.loc[[0,1], ['email', 'last']]  #loc uses labels
```
